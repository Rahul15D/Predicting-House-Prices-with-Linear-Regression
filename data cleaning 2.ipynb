Missing value imputation by Mean, median

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python


df = pd.read_csv(r'C:\Users\SONU\Downloads\train.csv')
```

```python
df.shape
```

```python
df.info()
```

```python
pd.set_option('display.max_columns', None) # to get the max rows and col
pd.set_option('display.max_rows', None)
df.isnull().sum()
```

```python
missing_value_per=df.isnull().sum()/df.shape[0] *100
missing_value_per
```

```python
missing_value_per_col=missing_value_per[missing_value_per > 20 ].keys() 
missing_value_per_col
```

```python
df2_drop_col= df.drop(columns = missing_value_per_col) 
df2_drop_col.shape
```

```python
df3_num=df2_drop_col.select_dtypes(include=['int64','float64'])
df3_num.head()
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(df3_num.isnull())
```

```python
df3_num[df3_num.isnull().any(axis=1)]
```

```python
df3_num.isnull().sum()
```

```python
missing_num_var = [ var for var in df3_num.columns if df3_num[var].isnull().sum()>0]
missing_num_var
```

```python
plt.figure(figsize=(10, 10))
sns.set()
# Loop through the variables and create subplots
for i, var in enumerate(missing_num_var):
    plt.subplot(2, 2, i + 1)
    
    # Overlay distogram
    sns.displot(df3_num[var], bins=20, kde_kws={'linewidth':5,'color':'#DC143C'})
```

```python
df4_mean=df3_num.fillna(df3_num.mean())
df4_mean.isnull().sum().sum()
```

```python
plt.figure(figsize=(10, 10))

# Loop through the variables and create subplots
for i, var in enumerate(missing_num_var):
    plt.subplot(2,2, i + 1)
    
    # Overlay histograms
    sns.histplot(df3_num[var], bins=20, color='g', kde=True, label='Original Data', alpha=0.5)
    sns.histplot(df4_mean[var], bins=20, color='r', kde=True, label='Filtered Data', alpha=0.5)
    
    plt.title(var)
    plt.legend()

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```

```python
df5_median=df3_num.fillna(df3_num.median())
df5_median.isnull().sum().sum()
```

```python
plt.figure(figsize=(10, 10))

# Loop through the variables and create subplots
for i, var in enumerate(missing_num_var):
    plt.subplot(2,2, i + 1)
    
    # Overlay histograms
    sns.histplot(df3_num[var], bins=20, color='g', kde=True, label='Original Data', alpha=0.5)
    sns.histplot(df4_mean[var], bins=20, color='r', kde=True, label='mean Data', alpha=0.5)
    sns.histplot(df5_median[var], bins=20, color='k', kde=True, label='median Data', alpha=0.5)
    
    plt.title(var)
    plt.legend()

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```
