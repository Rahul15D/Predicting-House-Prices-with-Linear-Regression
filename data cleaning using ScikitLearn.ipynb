# Numerical Missing value by class SimpleImput


```python
imort numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
```

```python
train = pd.read_csv(r'C:\Users\SONU\Downloads\train.csv')
test = pd.read_csv(r'C:\Users\SONU\Downloads\test.csv')
print("shape of train df =", train.shape)
print("shape of test df =", test.shape)
```

```python
X_train = train.drop(columns="SalePrice")
Y_train= train["SalePrice"]
print("shape of X_train df =", X_train.shape)
print("shape of Y_train df =", Y_train.shape)
```

# Numerical Missing Value Imputation

```python
num_vars=X_train.select_dtypes(include=["int64","float64"]).columns
num_vars
```

```python
X_train[num_vars].isnull().sum()  # for knowing the how many values  are missing
```

```python
imputer_mean = SimpleImputer(strategy= 'mean') # applying the mean term for puting the missung value 
```

```python
imputer_mean.fit(X_train[num_vars])
```

```python
imputer_mean.statistics_
```

```python
imputer_mean.transform(X_train[num_vars])
```

```python
X_train[num_vars] = imputer_mean.transform(X_train[num_vars])
test[num_vars] = imputer_mean.transform(test[num_vars])
```

```python
X_train[num_vars].isnull().sum()
```

# Categorial Missing Value Imputation

```python
car_vars=X_train.select_dtypes(include=["O"]).columns
car_vars
```

```python
X_train[car_vars].isnull().sum()
```

```python
imputer_mode = SimpleImputer(strategy= 'most_frequent')
imputer_mode
```

```python
imputer_mode.fit(X_train[car_vars])
```

```python
imputer_mode.statistics_
```

```python
X_train[car_vars] = imputer_mode.transform(X_train[car_vars])
test[car_vars] = imputer_mode.transform(test[car_vars])
```
