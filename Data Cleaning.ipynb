```python
null_var = df.isnull().sum()/df.shape[0]    * 100# for konow about the what % of data is dirty
null_var
```

```python
drop_columns = null_var[null_var > 17 ].keys()  # for know about the which col is deleting
drop_columns
```

```python
df2_drop_col= df.drop(columns = drop_columns) 
```

```python
df3_drop_rows=df2_drop_col.dropna() # for droping the nan value rows
```

```python
plt.figure(figsize=(8, 6))

# Overlay  histograms 
sns.histplot(df['MSSubClass'], color='g', label='Original Data')
sns.histplot(df3_drop_rows['MSSubClass'], color='r', label='Filtered Data')

plt.legend()
plt.title('Overlapping Distributions of MSSubClass')
plt.show()
```

```python
num_var2 = [ 'MSSubClass', 'LotArea', 'OverallQual', 'OverallCond',
       'YearBuilt', 'YearRemodAdd', 'MasVnrArea', 'BsmtFinSF1', 'BsmtFinSF2',
       'BsmtUnfSF', 'TotalBsmtSF', '1stFlrSF', '2ndFlrSF', 'LowQualFinSF',
       'GrLivArea', 'BsmtFullBath', 'BsmtHalfBath', 'FullBath', 'HalfBath',
       'BedroomAbvGr', 'KitchenAbvGr', 'TotRmsAbvGrd', 'Fireplaces',
       'GarageYrBlt', 'GarageCars', 'GarageArea', 'WoodDeckSF', 'OpenPorchSF',
       'EnclosedPorch', '3SsnPorch', 'ScreenPorch', 'PoolArea', 'MiscVal',
       'MoSold', 'YrSold', 'SalePrice']


plt.figure(figsize=(25, 25))

# using loop create  variables and create subplots
for i, var in enumerate(num_var2):
    plt.subplot(9, 4, i + 1)
    
    # Overlay histograms
    sns.histplot(df[var], bins=20, color='g', kde=True, label='Original Data', alpha=0.5)
    sns.histplot(df3_drop_rows[var], bins=20, color='r', kde=True, label='Filtered Data', alpha=0.5)
    
    plt.title(var)
    plt.legend()

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```

```python
def cat_var_dist(var):
   return pd.concat([df[var].value_counts()/df.shape[0] * 100,
          df3_drop_rows[var].value_counts()/df3_drop_rows.shape[0] * 100],axis=1,keys=[var+'_org',var+'clean'])
    
```
